```python
def rep(rawstr, dic_rep):
    for i in dic_rep:
        rawstr = rawstr.replace(i, dic_rep[i])
    return rawstr
```

```python
clean_dic = {'_x000D_':'', '\n':'@@@@', '\t':'@@@@', '. ':'@@@@'}
```

```python
main['general_comments_split'] = [rep(str(i), clean_dic).split('@@@@') for i in main.general_comments]
main['full_work_history_split'] = [rep(str(i), clean_dic).split('@@@@') for i in main.full_work_history]
main['description_claim_split'] = [rep(str(i), clean_dic).split('@@@@') for i in main.description_claim]
```

```python
def search(sentenses, lst, trunc=True):
    res = []
    for s in sentenses:
        for i in lst:
            if s.find(i)!=-1:
                res += [s]
                break
        if trunc and len(res)>=3:
            return res
    return res
#     return len(res)
```

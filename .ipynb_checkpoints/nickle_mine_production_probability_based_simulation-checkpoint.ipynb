# Data Processing

```python
df = pd.merge(main, nickel_mine_ma_proccessed, how='outer')
df.shape
df = df.dropna(subset=['resource (milion mt)', 'resv_metal_tonnage (mt)', 'production (mt)'], how='all')
df.shape
```

## production_capacity

```python
df['production_capacity_openpit'] = 0.123 * df['resv_ore_tonnage (mt)']**0.65 * df['grade (%)'] * 350
df['production_capacity_underground'] = 0.297 * df['resv_ore_tonnage (mt)']**0.56 * df['grade (%)'] * 350
df['production_capacity'] = df.apply(lambda x: x['production_capacity_openpit'] if x['mine_type1']=='Open Pit' else
                                     x['production_capacity_underground'] if x['mine_type1']=='Underground' else
                                     (x['production_capacity_openpit'] + x['production_capacity_underground'])/2, axis=1)
```

```python
df['production_capacity'] *= df['grade (%)']
```

```python
df['productiion_to_capacity'] = df['production (mt)'] / df['production_capacity']
df['productiion_to_capacity'].describe()
```

```python
df['mine_life'] = df['resv_metal_tonnage (mt)'] / df['production_capacity']
df['mine_life'].describe()
```

```python
df['mine_life'] = df['resv_metal_tonnage (mt)'] / df['production (mt)']
df['mine_life'].describe()
```

## regional risk and probability of shutdown

```python
df['country/region_ter_risk'] = df['country/region_ter_risk'].map({'Medium':2, 'High':3, 'Low':1, 'Insignificant':0}).fillna(2)
df['country/region_sec_risk'] = df['country/region_sec_risk'].map({'Medium':2, 'High':3, 'Low':1, 'Insignificant':0}).fillna(2)
df['country/region_pol_risk'] = df['country/region_pol_risk'].map({'Medium':2, 'High':3, 'Low':1, 'Insignificant':0}).fillna(2)
df['country/region_op_risk'] = df['country/region_op_risk'].map({'Medium':2, 'High':3, 'Low':1, }).fillna(2)
df['country/region_overall_risk'] = df['country/region_ter_risk'] + df['country/region_sec_risk'] + \
                                    df['country/region_pol_risk'] + df['country/region_op_risk']
df['prob_shutdown'] = np.power(df['country/region_overall_risk']/10, 6)
```

```python
df2 = df.copy()

year = 2020
production_list = [2500000]

exp_stage_index = df2['current_stage']=='Late-stage exploration'
df2.loc[exp_stage_index, 'late_exp_stage_lead_time'] = np.round(np.random.uniform(0, 3, size=sum(exp_stage_index)), 0)
df2.loc[exp_stage_index, 'late_exp_stage_left_time'] = np.round(df2.loc[exp_stage_index, 'late_exp_stage_lead_time'] * np.random.rand(sum(exp_stage_index)), 0)

fea_stage_index = df2['current_stage']=='Feasibility-stage'
df2.loc[fea_stage_index, 'fea_stage_lead_time'] = np.round(chi2.rvs(size=sum(fea_stage_index), **f1.get_best()['chi2']), 0)
df2.loc[fea_stage_index, 'fea_stage_left_time'] = np.round(df2.loc[fea_stage_index, 'fea_stage_lead_time'] * np.random.rand(sum(fea_stage_index)), 0)

dev_stage_index = df2['current_stage'].isin(['Development-stage', 'Construction-stage'])
df2.loc[dev_stage_index, 'dev_stage_lead_time'] = np.round(uniform.rvs(size=sum(dev_stage_index), **f2.get_best()['uniform']), 0)
df2.loc[dev_stage_index, 'dev_stage_left_time'] = np.round(df2.loc[dev_stage_index, 'dev_stage_lead_time'] * np.random.rand(sum(dev_stage_index)), 0)

tp = []
surplus_cont_count = 0
for i in range(1, len(demand_list)):
    year += 1
    
    prev_demand = demand_list[i-1]
    prev_production = production_list[i-1]
    cur_demand = demand_list[i]
    supply_surplus = prev_production - cur_demand
    supply_surplus_ratio = supply_surplus / cur_demand
    print(np.round(supply_surplus_ratio, 2))


    if supply_surplus_ratio < -0.01:
    # transit to next stage
        print('deficit')
        late_stage_finished_index = df2['late_exp_stage_left_time']<=0
        df2.loc[late_stage_finished_index, 'fea_stage_lead_time'] = np.round(chi2.rvs(size=sum(late_stage_finished_index), **f1.get_best()['chi2']), 0)
        df2.loc[late_stage_finished_index, 'fea_stage_left_time'] = np.round(df2.loc[late_stage_finished_index, 'fea_stage_lead_time'] * np.random.rand(sum(late_stage_finished_index)), 0)
        df2.loc[late_stage_finished_index, 'late_exp_stage_left_time'] = np.nan
        df2.loc[late_stage_finished_index, 'current_stage'] = 'Feasibility-stage'

        resv_delcare_index = df2['fea_stage_left_time']<=0
        df2.loc[resv_delcare_index, 'dev_stage_lead_time'] = np.round(uniform.rvs(size=sum(resv_delcare_index), **f2.get_best()['uniform']), 0)
        df2.loc[resv_delcare_index, 'dev_stage_left_time'] = np.round(df2.loc[resv_delcare_index, 'dev_stage_lead_time'] * np.random.rand(sum(resv_delcare_index)), 0)
        df2.loc[resv_delcare_index, 'fea_stage_left_time'] = np.nan
        df2.loc[resv_delcare_index, 'current_stage'] = 'Development-stage'

        dev_finished_index = df2['dev_stage_left_time']<=0
        print(sum(dev_finished_index))
        df2.loc[dev_finished_index, 'dev_stage_left_time'] = np.nan
        df2.loc[dev_finished_index, 'current_stage'] = 'Operating-stage'
        df2.loc[dev_finished_index, 'production_capacity_openpit'] = 0.123 * df2.loc[dev_finished_index, 'resv_ore_tonnage (mt)']**0.65 * df2.loc[dev_finished_index, 'grade (%)'] * 350
        df2.loc[dev_finished_index, 'production_capacity_underground'] = 0.297 * df2.loc[dev_finished_index, 'resv_ore_tonnage (mt)']**0.56 * df2.loc[dev_finished_index, 'grade (%)'] * 350
        df2.loc[dev_finished_index, 'production_capacity'] = df2.loc[dev_finished_index, ].apply(lambda x: x['production_capacity_openpit'] if x['mine_type1']=='Open Pit' else
                                             x['production_capacity_underground'] if x['mine_type1']=='Underground' else
                                             (x['production_capacity_openpit'] + x['production_capacity_underground'])/2, axis=1)
        df2.loc[dev_finished_index, 'production (mt)'] = df2.loc[dev_finished_index, 'production_capacity'] * df2['productiion_to_capacity'].median()

        op_stage_index = df2['current_stage']=='Operating-stage'
        df2.loc[op_stage_index, 'production (mt)'] *= (1-supply_surplus_ratio)
        
        surplus_cont_count=0

    elif supply_surplus_ratio>0.01:
        print('surplus')
        op_stage_index = df2['current_stage']=='Operating-stage'
#         print(df2.loc[op_stage_index, 'production (mt)'].sum())
        df2.loc[op_stage_index, 'production (mt)'] *= (1-supply_surplus_ratio)
#         print(df2.loc[op_stage_index, 'production (mt)'].sum())

        surplus_cont_count += 1

#         if surplus_cont_count==3:
            
# halt

    # exploration stage
    exp_stage_index = df2['current_stage']=='Late-stage exploration'
    df2.loc[exp_stage_index, 'late_exp_stage_left_time'] -= 1                                                         
    
    # feasibility stage
    fea_stage_index = df2['current_stage']=='Feasibility-stage'
    df2.loc[fea_stage_index, 'fea_stage_left_time'] -= 1

    # reserve delcaration
    resv_delcare_index = df2['fea_stage_left_time']==0
    modification_rate = np.random.normal(loc=0.4, scale=0.1, size=sum(resv_delcare_index))
    df2.loc[resv_delcare_index, 'resv_metal_tonnage (mt)'] = df2.loc[resv_delcare_index, 'resource nickel contained (milion mt)'] * modification_rate

#     development stage
    dev_stage_index = df2['current_stage'].isin(['Development-stage', 'Construction-stage'])
    df2.loc[dev_stage_index, 'dev_stage_left_time'] -= 1
    

    df2['production (mt)'] = df2.apply(lambda x: min(x['production (mt)'], x['production_capacity']), axis=1)
    df2['production (mt)'] = df2.apply(lambda x: min(x['production (mt)'], x['resv_metal_tonnage (mt)']), axis=1)
    df2['resv_metal_tonnage (mt)'] -= df2['production (mt)']
    production = df2['production (mt)'].sum()

    resv_dep_index = df2['resv_metal_tonnage (mt)']<=0
    df2.loc[resv_dep_index, 'resv_metal_tonnage (mt)'] = 0
    df2.loc[resv_dep_index, 'production (mt)'] = 0
    df2.loc[resv_dep_index, 'current_stage'] = 'Closed'
    df2.loc[resv_dep_index, 'actv_status'] = 'Inactive'

    production_list += [production]
#     tp += [df2['current_stage'].value_counts()]

#     col = ['prop_name', 'dev_stage', 'actv_status', 'country_name', 'current_stage', 'production_capacity', 'resv_metal_tonnage (mt)',
#            'grade (%)', 'production (mt)', 'resource (milion mt)', 'resource nickel contained (milion mt)', 'resv_ore_tonnage (mt)',
#            'late_exp_stage_lead_time', 'late_exp_stage_left_time', 'fea_stage_lead_time', 'fea_stage_left_time', 'dev_stage_lead_time',
#            'dev_stage_left_time', 'latitude', 'longitude']

    col = ['prop_name', 'dev_stage', 'actv_status', 'country_name', 'current_stage', 'production_capacity', 'resv_metal_tonnage (mt)',
           'grade (%)', 'production (mt)', 'resource (milion mt)', 'resource nickel contained (milion mt)', 'resv_ore_tonnage (mt)',
           'latitude', 'longitude']

#     tp += [df2[df2['prop_name']=='Alexo-Dundonald'][col]]
    tp += [df2[col]]
```
